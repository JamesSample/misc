## 1. Read data

```python
# Read data
df = pd.read_csv('hg_dep_2000-2013.csv')
df.head()
```

## 2. Map

```python
# Map of points
map1 = nivapy.spatial.quickmap(df, lon_col='Longitude', lat_col='Latitude', popup='JR-ID')
map1
```

## 3. Assign points to "regine" catchments

```python
# Path to Regine catchment shapefile
reg_shp_path = (r'./gis/reg_minste_f_wgs84.shp')

# Spatial join
df = identify_point_in_polygon(df, reg_shp_path, 
                               'JR-ID', 'VASSDRAGNR',
                               'Latitude', 'Longitude')

df.head()
```

```python
# Extract main catcment number
df['main_vassdrag'], _ = df['VASSDRAGNR'].str.split('.', 1).str
df['main_vassdrag'] = df['main_vassdrag'].astype(float)

df.head()
```

## 4. Assign to OSPAR region

```python
def assign_ospar(x):   
    if x in range(1, 24):
        return 'SKAGERAK'
    elif x in range(24, 91):
        return 'NORTH SEA'
    elif x in range(91, 171):
        return 'NORWEGIAN SEA2'
    elif x in range(171, 248):
        return 'LOFOTEN-BARENTS SEA'
    else:
        return np.nan

# Assign main catchments to OSPAR regions
df['osp_reg'] = df['main_vassdrag'].apply(assign_ospar)

df.head()
```

```python
df.to_csv('hg_dep_2000-2013_ospar.csv')
```
